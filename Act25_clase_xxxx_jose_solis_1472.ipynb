```python
class Publicacion:
    def __init__(self, titulo, contenido, autor):
        self.titulo = titulo
        self.contenido = contenido
        self.autor = autor

    def leer_datos(self):
        print(f"\n[Sistema]: Has abierto el archivo: {self.titulo}")
        print(f"Texto: {self.contenido}")

    def mostrar_datos(self):
        print(f"\n--- FICHA: {self.titulo} ---")
        print(f"Escrito por: {self.autor}")
        print("----------------------------")

# Creamos los dos objetos iniciales
post1 = Publicacion("Aprender Python", "La POO es útil para organizar código.", "Gemini")
post2 = Publicacion("Receta de Cocina", "Ingredientes: Harina, huevos, leche.", "Chef AI")

# Menú interactivo
while True:
    opcion = input("\n¿Qué post quieres consultar? (1: post1, 2: post2, 0: salir): ")

    if opcion == "1":
        post1.mostrar_datos()
        if input("¿Deseas leer el contenido? (s/n): ").lower() == 's':
            post1.leer_datos()
    elif opcion == "2":
        post2.mostrar_datos()
        if input("¿Deseas leer el contenido? (s/n): ").lower() == 's':
            post2.leer_datos()
    elif opcion == "0":
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida.")

print("programa realizado por jose solis NL 51")
```
